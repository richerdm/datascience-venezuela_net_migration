```python
hist_indicator='Net migration'
hist_country='VEN'

mask1 = indicators['IndicatorName'].str.contains(hist_indicator)
mask2 = indicators['CountryCode'].str.contains(hist_country)

stage = indicators[mask1 & mask2]
stage.head(100)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8), dpi=80)

plt.ylabel('Net migration')
plt.xlabel('Year')

X = stage['Year']
Y = stage['Value']

plt.scatter(X,Y)
plt.plot(X,Y)

plt.show()
```

```python
hist_indicator='Net migration'
hist_country_LA='LCN'
hist_country_name_LA='Latin'

mask1_LA = indicators['IndicatorName'].str.contains(hist_indicator)
mask2_LA = indicators['CountryCode'].str.contains(hist_country_LA)
mask3_LA = indicators['CountryName'].str.contains(hist_country_name_LA)

stage_LA = indicators[mask1_LA & mask2_LA]
stage_LA.head(100)
```

```python
plt.figure(figsize=(12, 8), dpi=80)

plt.ylabel('Net migration')
plt.xlabel('Year')

X = stage['Year']
Y = stage['Value']

X1 = stage_LA['Year']
Y1 = stage_LA['Value']/46

plt.plot(X,Y,'o-',label='Venezuela')

plt.plot(X1,Y1,'o-',label='Latin America and the Caribbean (average of the 46 countrys)')

plt.title('Net Migration')

plt.legend()

plt.show()
```
